```
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv')

df.head()
```

```
print (df.dtypes)
```

```
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```
df[['State', 'Utilization Type', 'Product Name', 'Total Amount Reimbursed', 'Quarter Begin', 'Quarter Begin Date', 'Location' ]] = df[['State', 'Utilization Type', 'Product Name', 'Total Amount Reimbursed', 'Quarter Begin', 'Quarter Begin Date', 'Location' ]].astype('category')

print(df.info())
df.head()
```

```
df['Total Amount Reimbursed'] = df['Total Amount Reimbursed'].astype('str')

print(df.info())
df.head()
```

```
print (df.dtypes)
```

```

df['Units Reimbursed'] = df['Units Reimbursed'].astype('str')
print (df.dtypes)
```
